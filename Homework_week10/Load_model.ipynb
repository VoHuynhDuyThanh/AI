```python
from keras.models import load_model
model = load_model("/content/drive/MyDrive/Colab Notebooks/Cifar100.h5")
```

```python
import matplotlib.pyplot as plt
from keras.utils import load_img, img_to_array
import numpy as np

img = load_img('/content/drive/MyDrive/AI/apple.jpeg', target_size = (32,32))
plt.imshow(img)
img = img_to_array(img)
img = img.reshape(1, 32*32*3)
img = img.astype('float32')
img = img / 255
np.argmax(model.predict(img),axis = -1)
```

```python
from matplotlib.image import imread
folder = '/content/drive/MyDrive/AI/Cifar100'
for i in range(1,20):
  plt.subplot(6,6,i+1)
  filename = folder +'/'+'A' +str(i)+'.jpeg'
  img = imread(filename)
  plt.imshow(img)
plt.show()
```
