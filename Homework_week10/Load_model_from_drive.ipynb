```python
from keras.models import load_model
```

```python
model100 = load_model('/content/drive/MyDrive/AI/AI_Backup/Cifar100CNN.h5')
```

```python
from keras.utils import load_img, img_to_array
from matplotlib import pyplot as plt
url = '/content/drive/MyDrive/AI/Cifar100/A1.jpeg'
img = load_img(url)
plt.imshow(img)
plt.show()
```

```python
img = load_img(url, target_size= (32,32))
plt.imshow(img)
```

```python
img = img_to_array(img)
img = img.reshape(1,32,32,3)
img = img.astype('float32')/255

import numpy as np
np.argmax(model100.predict(img), axis = 1)
```
