```python
import pandas as pd

url = 'mtcars.csv'
data = pd.read_csv(url)

data.hist()
data.corr()

X = data.drop(data. columns[[0,2]], axis = 1)
data.plot(kind = 'scatter', x ='wt', y ='mpg', figsize = (9,9), color = 'black')
```

```python
from sklearn import linear_model
X = pd.DataFrame(data['wt'])
y = data['mpg']

model = linear_model.LinearRegression()
model.fit(X, y) 
print(model.intercept_)
print(model.coef_)
model.score(X,y)
```

```python
y_pred = model.predict(X)
residual = (data['mpg'] - y_pred)
residual.describe()
```

```python
data.plot(kind = 'scatter', x ='wt', y ='mpg', figsize = (9,9), color = 'black')
plt.plot(data['wt'], y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt'] **2]).T    # a1x + a2x^2 + b, .T la nghich dao
model.fit(X, y = data['mpg'])
print('Intercept ca m hnh bc 2: ')
print(model.intercept_)
print('Coeffient ca m hnh bc 2: ')
print(model.coef_)
print(' chnh xc ca m hnh: ')
print(model.score(X, y))
```

```python
range = np.arange(1.5, 5.5, 0.1)
predictor = pd.DataFrame([range, range **2]).T
y_pred = model.predict(X = predictor)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize = (9, 9), 
          color = 'red', xlim = (0, 7))
plt.plot(range, y_pred, color = 'blue')
```
